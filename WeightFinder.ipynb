```python
import pandas as pandas
import numpy as numpy

# get counts
trainingLabels = pandas.read_csv('A. RFMiD_All_Classes_Dataset\\a. RFMiD_Training_Labels.csv')

# the columns that count the diseases
diseaseCols = trainingLabels.columns[2:]

#count of all disease incidents, both in total in their own columns as well
counts = trainingLabels[diseaseCols].sum()

#calculates the Frequencies of each disease vs the number of patients
frequencies = counts / len(trainingLabels)

# finds the "Middle" (Median) Frequency
median_freq = frequencies.median()

# Calculate the Weights
weights = median_freq / frequencies

print("--- Median Frequency Weights ---")
print(weights)
```
